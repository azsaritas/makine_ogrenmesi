## Karar Ağacı (Decision Tree) Nedir?

Karar ağaçları, makine öğreniminde ve karar verme süreçlerinde kullanılan bir modelleme yöntemidir. Bir karar ağacı, veriyi sınıflandırmak veya tahmin yapmak için dallanan bir yapıdır. Bu yapı, bir dizi karar kurallarını temsil eder ve bu kurallar, veriyi belirli kriterlere göre dallara ayırarak ilerler. Karar ağaçları, her bir düğümde bir "karar" veya "koşul" sunar ve bu koşula göre veri, bir sonraki dala yönlendirilir.

#### Karar Ağaçlarının Temel Bileşenleri
Kök Düğüm (Root Node): Ağacın başlangıç noktasıdır ve tüm veriyi içerir. Burada, veriyi ilk defa bölmek için kullanılan karar bulunur.

İç Düğümler (Internal Nodes): Kök düğümden gelen dallar üzerinde yer alan düğümlerdir. Her iç düğüm, belirli bir özelliğe göre veri kümesini böler.

Yaprak Düğümler (Leaf Nodes): Veri setinin son sınıflandırıldığı veya tahmin yapıldığı düğümlerdir. Bu düğümler, daha fazla dalı olmayan düğümlerdir.

#### Karar Ağaçlarının Kullanım Alanları
Sınıflandırma: Verilen veri örneklerini önceden tanımlanmış kategorilere ayırmak için kullanılır. Örneğin, bir e-posta'nın spam olup olmadığını belirlemek.

Regresyon: Sürekli değerler tahmin etmek için kullanılır. Örneğin, bir evin fiyatını tahmin etmek.

Veri Analizi: Verinin yapısını ve ilişkilerini görselleştirmek ve anlamak için kullanılır.

#### Avantajları
Kolay Anlaşılabilirlik: Karar ağaçları görselleştirilebilir ve sonuçları kolayca yorumlanabilir.

Esneklik: Hem sınıflandırma hem de regresyon problemlerinde kullanılabilir.

Az Veri Ön İşleme: Karar ağaçları, verideki eksik değerlerle ve veri ön işleme gereksinimlerinin az olmasıyla başa çıkabilir.

#### Dezavantajları
Aşırı Uyum (Overfitting): Karar ağaçları, özellikle derin ağaçlarda, eğitim verisine çok iyi uyum sağlayarak genel performansı düşürebilir.

Kararsızlık: Küçük veri değişiklikleri, ağacın yapısını önemli ölçüde değiştirebilir.

#### Karar ağacı tipleri ikiye ayrılır:
• Entropiye dayalı sınıflandırma ağaçları (ID3, C4.5)
• Regresyon ağaçları (CART).

Karar ağaçları çok boyutlu (özellikli) veriyi belirlenmiş şartlara bağlı olarak parçalara böler. 
Her adımda verinin hangi özelliği üzerinde işlem yapılacağına karar vermek çok büyük bir kombinasyonun çözümüyle mümkündür. Örneğin, 5 özellik ve 20 örneğe sahip bir veride 10^6 dan fazla sayıda farklı karar ağacı oluşturulabilir. Bu sebeple her parçalanmanın metodolojik olması gerekir. Quinlan’e göre veri bir özelliğine göre bölündüğünde elde edilen her bir veri kümesinin belirsizliği minimum ve dolayısıyla bilgi kazancı maksimum ise en iyi seçim yapılmıs demektir. Buna göre önerdiği ilk algoritma ID3’te tek tek özellik vektörleri incelenir ve en yüksek bilgi kazancına sahip özellik, ağaçta dallanma yapmak için tercih edilir.

### Karar Ağacı Algoritmasının Adımları

#### Özellik Seçimi ve Bölme:

Gini Katsayısı veya Entropi Hesaplama: Algoritma, hangi özelliğin veri setini en iyi şekilde böleceğine karar verir. Bunun için genellikle iki ana ölçüt kullanılır: Gini katsayısı (Gini impurity) ve entropi (information gain).
Bilgi Kazancı (Information Gain): Hangi özelliğin veri setindeki belirsizliği en çok azalttığını ölçer. Entropi veya Gini katsayısı kullanılarak bilgi kazancı hesaplanır.

#### Dallanma (Splitting):

Seçilen özelliğe göre veri seti dallara ayrılır. Bu süreç, her özellik ve her olası bölme noktası için tekrarlanır. En yüksek bilgi kazancına sahip özellik ve bölme noktası seçilir.

#### Düğüm Oluşturma:

İç Düğümler (Internal Nodes): Her iç düğüm, veri setini seçilen özelliğe ve bölme noktasına göre iki veya daha fazla alt küme ayırır.
Yaprak Düğümler (Leaf Nodes): Verinin tamamen sınıflandırıldığı veya tahmin yapıldığı son düğümlerdir. Bu düğümler daha fazla dallanmaz.

#### Tekrarlama:

İç düğümler oluşturulduktan sonra, aynı süreç her bir alt küme için tekrarlanır. Bu, ağacın tüm veri setini en iyi şekilde temsil edecek şekilde dallanmasına kadar devam eder.

#### Durdurma Kriterleri:

Ağacın dallanmasını durdurmak için belirli kriterler kullanılır. Örneğin, bir düğümdeki veri sayısı belirli bir eşikten azsa, ağacın derinliği belirli bir seviyeye ulaştıysa veya bilgi kazancı belirli bir seviyenin altındaysa dallanma durdurulur.

#### Budama (Pruning):

Ağaç büyüdükten sonra, aşırı uyumu önlemek için budama yapılabilir. Budama, ağaçtaki gereksiz dalları kaldırarak daha basit ve genelleştirilebilir bir model oluşturmayı amaçlar. Budama, eğitim ve doğrulama veri setleri kullanılarak yapılabilir.

## Karar Ağacı Örnek Uygulama

Bu örnekte çiçekleri, taç ve çanak yaprak boyutlarına göre sınıflandırıyoruz.

```python
from sklearn.datasets import *
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

import dtreeviz
import graphviz.backend as be
from IPython.display import Image, display_svg, SVG
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files/Graphviz/bin'
```

Gerekli kütüphaneleri ve uygulama yollarını import ettikten sonra veri setini yüklüyoruz.

```python
iris = load_iris() #iris veri seti

X_train = iris.data
y_train = iris.target
```

```python
clas = DecisionTreeClassifier() #kullanacağımız sınıflandırma algoritması
clas.fit(X_train, y_train)
```

```python
observation = [[ 5, 4, 3, 2]] # gözlem sınıfının tahmini
clas.predict(observation)
clas.predict_proba(observation)
```

Karar ağacının yapısını çizdiriyoruz.

```python
plot_tree(clas)
```

```python
viz = dtreeviz.model(clas, X_train, y_train, feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"])
viz
```

Bu adımda dtreeviz paketinden içe aktarılacak Image sınıfını kullanarak karar ağacını görselleştiriyoruz.

```python
viz.view(fancy=False)
```

Oluşturulan bu karar ağacı bize çiçeklerin yapraklarına göre sınıflandırıyor.

```python
viz.view(scale=1)
```
